```python
#    object  datetime
df['Datetime'] = pd.to_datetime(df['Datetime'])
```

```python
df.info()
```

```python
#   -   
df['date'] = [d.date() for d in df['Datetime']]
df['time'] = [d.time() for d in df['Datetime']]
```

```python
#       
df['PC_Sum']=df["PowerConsumption_Zone1"]+df["PowerConsumption_Zone2"]+df["PowerConsumption_Zone3"]
```

```python
df.head(3)
```

```python
#      
df_data = ['date', 'time', 'Temperature', 'PC_Sum']
```

```python
df_data=df[df_data]
df_data.head(3)
```

```python
#      
df_date_mean = df_data.groupby(['date']).mean()
df_date_mean
```

```python
g=sns.PairGrid(df_date_mean)
g.map_diag(sns.histplot, bins = 10)
g.map_offdiag(sns.scatterplot)
```

```python
#      
df_date_min = df_data.groupby(['date']).min()
df_date_min
```

```python
g=sns.PairGrid(df_date_min)
g.map_diag(sns.histplot, bins = 10)
g.map_offdiag(sns.scatterplot)
```

```python
#      
df_date_max = df_data.groupby(['date']).max()
df_date_max
```

```python
g=sns.PairGrid(df_date_max)
g.map_diag(sns.histplot, bins = 10)
g.map_offdiag(sns.scatterplot)
```

```python
sns.scatterplot(data = df_date_mean, x = "Temperature", y = "PC_Sum", hue = "date")
plt.grid()
plt.legend(["PowerConsumption_mean"]);
```

```python
df_date_mean.describe()
```

```python
#      
df_time_mean = df_data.groupby(['time']).mean()
df_time_mean
```

```python
#      
df_time_min = df_data.groupby(['time']).min()
df_time_min
```

```python
#      
df_time_max = df_data.groupby(['time']).max()
df_time_max
```

```python
fig, axes = plt.subplots(2,3) 

p0_0 = axes[0,0].scatter(data = df_date_mean, x = "Temperature", y = "PC_Sum", c = "r")     
axes[0,0].grid(color = 'white')
axes[0,0].set_facecolor("lavender")
axes[0,0].set_xlabel('temperature')
axes[0,0].set_ylabel('powerconsumption_mean')

p0_1 = axes[0,1].scatter(data = df_date_min, x = "Temperature", y = "PC_Sum", c = "r")                               
axes[0,1].grid(color = 'white')
axes[0,1].set_facecolor("lavender")
axes[0,1].set_xlabel('temperature')
axes[0,1].set_ylabel('powerconsumption_min')

p0_2 = axes[0,2].scatter(data = df_date_max, x = "Temperature", y = "PC_Sum", c = "r")                               
axes[0,2].grid(color = 'white')
axes[0,2].set_facecolor("lavender")
axes[0,2].set_xlabel('temperature')
axes[0,2].set_ylabel('powerconsumption_max')

p1_0 = axes[1,0].scatter(data = df_time_mean, x = "Temperature", y = "PC_Sum", c = "blue")     
axes[1,0].grid(color = 'white')
axes[1,0].set_facecolor("lavender")
axes[1,0].set_xlabel('temperature')
axes[1,0].set_ylabel('powerconsumption_mean')

p1_1 = axes[1,1].scatter(data = df_time_min, x = "Temperature", y = "PC_Sum", c = "blue")                               
axes[1,1].grid(color = 'white')
axes[1,1].set_facecolor("lavender")
axes[1,1].set_xlabel('temperature')
axes[1,1].set_ylabel('powerconsumption_min')

p1_2 = axes[1,2].scatter(data = df_time_max, x = "Temperature", y = "PC_Sum",  c = "blue")                               
axes[1,2].grid(color = 'white')
axes[1,2].set_facecolor("lavender")
axes[1,2].set_xlabel('temperature')
axes[1,2].set_ylabel('powerconsumption_max')

fig.set_size_inches(30,15)
plt.show()
```
